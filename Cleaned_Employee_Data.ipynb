```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
data = pd.read_csv("Employee2.csv")
data
```

```python
data.describe()
```

```python
data.columns = data.columns.str.strip()
data["Salary"] = data["Salary"].replace('[\$,]', '', regex=True).astype(float)
print(data["Salary"].dtype)
data
```

```python
print(data.isnull().sum())
```

```python
print(data.duplicated().sum())
```

```python
data["HireDate"] = pd.to_datetime(data["HireDate"], errors="coerce")
data["FirstName"] = data["FirstName"].str.strip().str.capitalize()
data["LastName"] = data["LastName"].str.strip().str.capitalize()
data
```

```python
Q1 = data["Age"].quantile(0.25)
Q3 = data["Age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data["Age"] < lower_bound) | (data["Age"] > upper_bound)]

print("Number of outliers in Age:", len(outliers))
print(outliers[["EmployeeID", "Age"]])
numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    print(f"Column: {col}")
    print(f"Outliers count: {len(outliers)}")
    if len(outliers) > 0:
        print(outliers[[col]].head())  
    print("-"*40)
```

```python
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

data = data[(data['Age'] >= lower_limit) & (data['Age'] <= upper_limit)]
```

```python
Q1 = data['Salary'].quantile(0.25)
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

data['Salary'] = data['Salary'].clip(lower=lower_limit, upper=upper_limit)
```

```python
Q1 = data["Age"].quantile(0.25)
Q3 = data["Age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data["Age"] < lower_bound) | (data["Age"] > upper_bound)]

print("Number of outliers in Age:", len(outliers))
print(outliers[["EmployeeID", "Age"]])
numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    print(f"Column: {col}")
    print(f"Outliers count: {len(outliers)}")
    if len(outliers) > 0:
        print(outliers[[col]].head())  
    print("-"*40)
```

```python
data.to_csv("Cleaned_Employee_Data.csv", index=False)

data.to_excel("Cleaned_Employee_Data.xlsx", index=False)

print("✅ File saved successfully!")
```
